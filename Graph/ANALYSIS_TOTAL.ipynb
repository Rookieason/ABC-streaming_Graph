```python
# figure 10
index_selected = [31]
dist_list = ['ONLYLAB', 'LABV1']
path = pathlib.Path("../DASH-setup/client/logs/ubuntu18_trace9")
throughput_list = [1000]
idx=31
```

```python
df = pd.read_csv('../public_trace/Lumos5G/Lumos5G-v1.0.csv')
file_selected = []
for n, i in enumerate(index_selected):
    trace = df.loc[df['run_num']==i+1]
    sce = trace.iloc[0]["mobility_mode"]
    key = str(i)+'_'+sce
    file_selected.append(key)
# parameters
video_list = ['NA']
interface_list = ['mmwave']
abr_list = ['bolaEWMA10000100']
add1_list = ['nofast']
add2_list = ['abandon']
partition = 1
```

```python
book = {}
trace_set = {}
for v in video_list:
    for i in interface_list:
        for t in throughput_list:
            for d in dist_list:
                for abr in abr_list:
                    for add1 in add1_list:
                        for add2 in add2_list:
                            k = v + '_' + i + '_' + str(t) + '_' + d + '_' + abr + '_' + add1 + '_' + add2
                            book[k] = [{} for _ in range(partition)]
                            trace_set[k] = set()
                            for c, fs in enumerate(file_selected):
                                tmp = list(path.glob(f"*{v}*{i}*{t}*{fs}*_{d}_{abr}_{add1}*_{add2}*"))
                                if len(tmp):
                                    f = str(tmp[0])
                                    print(f)
                                    trace_set[k].add(str(f).split("half_")[-1].split('_')[0])
                                    index = int(c // (len(file_selected) / partition))
                                    book[k][index][fs] = {}
                                    tmp = open(f'{f}/qualityLog.json')
                                    book[k][index][fs]['quality'] = pd.read_json(tmp)
                                    tmp = open(f'{f}/segmentLog.json')
                                    book[k][index][fs]['segment'] = pd.read_json(tmp)
```

```python
#Figure 9 & Figure 10
dist = ['ONLYLAB', 'LABV1']

#Figure 10
# thr = ['1000']

#Figure 9
thr = ['4000']

name = ['LAB(original Abandon)', 'LAB(LAB Abandon)']
dim = ['qualityLevel', 'bufferLevel']
trace = df.loc[df['run_num']==idx+1]
key = str(idx)+'_'+trace.iloc[0]["mobility_mode"]
fig = plt.figure(figsize=(16, 12))
length = 0
for i, dimen in enumerate(dim):
    if dimen == 'qualityLevel':
        plt.subplot(311)
    if dimen == 'bufferLevel':
        plt.subplot(312)
    for t in thr:
        for d in dist:
            nm = [i + '(' + t + ')' for i in name]
            dictionary = 'NA_mmwave_' + t + '_' + d + '_bolaEWMA10000100_nofast_abandon'
            
            #Figure 10
#             interval = book[dictionary][0][key]['quality'][dimen][1500:3000]
            
            #Figure 9
            interval = book[dictionary][0][key]['quality'][dimen][3000:6500]
            
            plt.plot(interval)
            if dimen == 'qualityLevel':
                plt.ylabel('qualityLevel')
            else:
                plt.ylabel('bufferLevel(s)')
            plt.xlabel('Time')
            length = len(book[dictionary][0][key]['quality'][dimen])
            plt.legend(nm)
```
